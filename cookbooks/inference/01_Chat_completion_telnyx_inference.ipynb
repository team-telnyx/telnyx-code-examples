# First Chat Completion with Telnyx Inference

In this notebook, you’ll send your first chat completion request to a Telnyx-hosted model using the OpenAI Python SDK.

Telnyx Inference is OpenAI-compatible, so if you already use the OpenAI SDK, you can point it at Telnyx by changing the `base_url` and using your Telnyx API key.


## Install SDK

```python
!pip install openai
```

## Add Telnyx API key

```python
from getpass import getpass

TELNYX_API_KEY = getpass("Enter your Telnyx API key: ")
```

## Create Client
Telnyx Inference is OpenAI-compatible. This notebook uses the OpenAI Python SDK and points it at Telnyx with the `base_url` below.

```python
from openai import OpenAI

client = OpenAI(
    api_key=TELNYX_API_KEY,
    base_url="https://api.telnyx.com/v2/ai/openai"
)
```

## First Chat Completion

```python
response = client.chat.completions.create(
    model="moonshotai/Kimi-K2.6",
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant."
        },
        {
            "role": "user",
            "content": "Explain Telnyx Inference in one short paragraph."
        }
    ],
    temperature=0.2,
)

print(response.choices[0].message.content)
```

## Try Your Own Prompt

Change the prompt below and run the cell again.


```python
user_prompt = "Give me three example use cases for low-latency LLM inference."

response = client.chat.completions.create(
    model="moonshotai/Kimi-K2.6",
    messages=[
        {"role": "system", "content": "You are a concise technical assistant."},
        {"role": "user", "content": user_prompt}
    ],
    temperature=0.2,
)

print(response.choices[0].message.content)
```

## Common Errors

Here are a few common issues you may run into when making your first request.


```python
from openai import AuthenticationError, NotFoundError, RateLimitError, APIError

try:
    response = client.chat.completions.create(
        model="moonshotai/Kimi-K2.6",
        messages=[
            {"role": "user", "content": "Say hello in one sentence."}
        ],
        temperature=0.2,
    )

    print(response.choices[0].message.content)

except AuthenticationError:
    print("Authentication failed. Check that your Telnyx API key is correct.")

except NotFoundError:
    print("Model not found. Check that the model name is spelled correctly.")

except RateLimitError:
    print("Rate limit hit. Wait a moment and try again.")

except APIError as e:
    print(f"API error: {e}")
```
